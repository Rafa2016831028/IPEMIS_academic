```python
# Import necessary libraries
import pandas as pd
from google.colab import files
import numpy as np
from datetime import datetime, timedelta
import random

# Upload the file
uploaded = files.upload()


# Get the uploaded file name
file_name = list(uploaded.keys())[0]

# Read the file into a pandas DataFrame
df = pd.read_csv(file_name)

# Display the first few rows of the DataFrame
print("Preview of the uploaded file:")
print(df.head())
```

```python
# Check data types, null values, and basic statistics
print(df.info())
print(df.describe())
```

```python
def random_date(start, end):
    """
    Generate a random datetime between two datetime objects.
    """
    delta = end - start
    random_days = np.random.randint(0, delta.days + 1)  # Random number of days
    return start + timedelta(days=random_days)
```

```python
# Define the start and end dates
start_date = datetime(2018, 1, 1)
end_date = datetime(2021, 12, 31)

# Generate random dates for each row in the DataFrame
df['date'] = [random_date(start_date, end_date) for _ in range(len(df))]

# Display the DataFrame
print(df)
```

```python
random_values = ['Barishal',  'Khulna','Sylhet','Mymensingh', 'Rajshahi']

# Add a new column with random choices from the list
# df['from'] = np.random.choice(random_values, size=len(data))
df['from'] = ['Dhaka'] * 549 + ['Chittagong'] * 356 + ['Rangpur'] * 110 + list(np.random.choice(random_values, size=len(df) - 1015))

# Display the DataFrame
print(df)
```

```python
random_values = ['Class_1',  'Class_2','Class_3','Class_4']

# Add a new column with random values from the list
df['class'] =  [random.choice(random_values) for _ in range(len(df))]

print(df)
```

```python
random_values = ['Scholarship', 'Endorsement', 'Government Resource']

# Corresponding weights (biases)
weights = [0.3, 0.15, 0.55]  # Higher weight means higher chance

# Generate a single biased random choice
biased_choice = random.choices(random_values, weights=weights, k=1)[0]
print(f"Biased random choice: {biased_choice}")

# Generate multiple biased random choices
biased_choices = random.choices(random_values, weights=weights, k=10)
print(f"10 Biased random choices: {biased_choices}")

# Add a new column with random values from the list
df['Subscription_type'] =  [random.choices(random_values, weights=weights) for _ in range(len(df))]

print(df)
```
